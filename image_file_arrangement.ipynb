```python
# Rename the Image

def rename_images(directory, new_name_prefix):
    # List all files in the directory
    files = os.listdir(directory)
    
    # Filter out only image files (assuming they have extensions like .jpg, .jpeg, .png)
    image_files = [f for f in files if f.endswith(('.jpg', '.jpeg', '.png'))]
    
    # Loop over the image files and rename them
    for i, filename in enumerate(image_files):
        # Generate new name
        new_name = f"{new_name_prefix}_{i + 10000}.png"
        
        # Get full paths
        old_file = os.path.join(directory, filename)
        new_file = os.path.join(directory, new_name)
        
        # Rename the file
        os.rename(old_file, new_file)
```

```python
train_path = r'cifar10\cifar10\cifar10\train'

for folder in os.listdir(train_path):
    directory = os.path.join(train_path, folder)
    new_name_prefix = folder
    rename_images(directory, new_name_prefix)
```

```python
val_path = r'cifar10\cifar10\cifar10\test'

for folder in os.listdir(val_path):
    directory = os.path.join(val_path, folder)
    new_name_prefix = folder 
    rename_images(directory, new_name_prefix)
```

```python
train_data_ = {'Image_path':[], 'Label':[]}
train_path = r'cifar10\cifar10\cifar10\train'
for folder in os.listdir(train_path):
    image_path = os.path.join(train_path, folder)
    for img_path in os.listdir(image_path):
        train_data_['Image_path'].append(img_path)
        train_data_['Label'].append(folder)
train_annot = pd.DataFrame(train_data_)
train_annot
```

```python
val_data_ = {'Image_path':[], 'Label':[]}
val_path = r'cifar10\cifar10\cifar10\test'

for folder in os.listdir(val_path):
    image_path = os.path.join(val_path, folder)
    for img_path in os.listdir(image_path):
        val_data_['Image_path'].append(img_path)
        val_data_['Label'].append(folder)
val_annot = pd.DataFrame(val_data_)
val_annot
```

```python
classes = train_annot.Label.unique()
classes
```

```python
class_map = {clss:id for id, clss in enumerate(classes)}
class_map
```

```python
train_annot['Label'] = train_annot['Label'].map(class_map)
val_annot['Label'] = val_annot['Label'].map(class_map)
```
